```
file = requests.get('https://ocw.mit.edu/ans7870/6/6.006/s08/lecturenotes/files/t8.shakespeare.txt')
data=file.text.split('\n')
shak_data=data[253:]
len(shak_data)
shak_data= ''.join(shak_data)
```

```
from nltk.tokenize import sent_tokenize, word_tokenize

import nltk
nltk.download('punkt')
```

```
puncuations = nltk.word_tokenize(string.punctuation)
def preprocess(df):
    df= nltk.word_tokenize(df)
    tokens= [q.lower() for q in df if q not in puncuations and q.isalpha()]
    return tokens
tokens =preprocess(shak_data)
```

```
uniq_words=set(tokens)
l= len(tokens)
predict_len= 12+1
sent= list()
for i in range (predict_len, l):
    seq= tokens[i-predict_len:i]
    sent.append(' '.join(seq))
    if  i>20000: break
        # i set as conda going out of range 

print(sent)
```
